# Sample Superstore Data Visualization
This notebook shows some simple data visualizations like sales distribution, profit by category, and sales vs profit.

```python
from google.colab import drive
import os

# create the folder where Drive will be mounted
os.makedirs("/content/drive", exist_ok=True)

# mount Google Drive
drive.mount("/content/drive")

!ls /content/drive/MyDrive

import pandas as pd

file_path = '/content/drive/MyDrive/SampleSuperstore.csv.zip'

df = pd.read_csv(file_path)
df.head(20)
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
# 1 Sales distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Sales'], bins=30, kde=True)
plt.title("Sales Distribution")
plt.show()

# 2 Profit by Category
plt.figure(figsize=(8,5))
sns.barplot(x='Category', y='Profit', data=df)
plt.title("Profit by Category")
plt.show()

# 3 Sales vs Profit
plt.figure(figsize=(8,5))
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title("Sales vs Profit Relationship")
plt.show()
```
